### Creating a histogram of order_time

```python
# Creating a subset with only the columns of interest

order_time = instacart[['order_time', 'order_id']]
```

```python
order_time.head()
```

```python
# Creating a subset withou duplicates

order_time_nodups = order_time.drop_duplicates()
```

```python
order_time_nodups.head()
```

```python
histogram_order_time = order_time_nodups['order_time'].plot.hist(bins = 24)
plt.title('Distribution of Orders by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```

```python
# exporting plot

histogram_order_time.figure.savefig(os.path.join(path, 'Analysis','Final Visualization', 'histogram_order_time.png'))
```

### Creating a line chart with 'order_cost' x order_time = hourly expenditure

```python
# Creating a subset with only the columns of interest

order_cost_by_time = instacart[['order_time', 'order_id', 'order_cost']]
```

```python
order_cost_by_time.shape
```

```python
# Creating a subset withou duplicates

order_cost_by_time_no_dupes = order_cost_by_time.drop_duplicates()
```

```python
order_cost_by_time_no_dupes.shape
```

```python
line_order_cost_by_time = sns.lineplot(data = order_cost_by_time_no_dupes, x = 'order_time',y = 'order_cost')
plt.title('Variation of Cost of Orders by Hour of Day')
plt.xlabel('Order Time')
plt.ylabel('Order Cost')
plt.grid(True)

plt.show()
```

```python
# exporting plot

line_order_cost_by_time.figure.savefig(os.path.join(path, 'Analysis','Final Visualization', 'line_order_cost_by_time.png'))
```
